```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Sine lookup table of 30 values
t = np.linspace(0,2*np.pi,30)
sinvals = np.sin(t)
sinvals = 128*(1+sinvals) # Translating the values form -1 to 1 to 0 to 255
sinvals = sinvals.astype(int)
plt.plot(t,sinvals)
#sinvals = sinvals - int(np.mean(sinvals))
len(sinvals)
```

```python
# A bit of wavefrom representation
t = np.linspace(0,50*2*np.pi, 1000)
t_chunk = np.linspace(0,1*2*np.pi, 200)

wave_LD = np.array([])
for i in range(5):
    wave_LD = np.concatenate((wave_LD, np.sin(t_chunk)/5*(i+1)))
plt.plot((t/300),np.sin(t),label='Re-lock wave piezo')
plt.plot((t/300),wave_LD, label='Re-lock wave LD')    
plt.legend()
plt.xlabel('t (seconds)')
plt.ylabel('ampltide (arb.u.)')
```
